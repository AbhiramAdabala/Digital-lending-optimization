````python
# =============================================================================
# CELL 2 — STEP 1: Research Thesis Framework
# =============================================================================

section("STEP 1 — RESEARCH THESIS FRAMEWORK")

THESIS = """
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
  DIGITAL LENDING PORTFOLIO OPTIMIZATION:
  AN ENTERPRISE AI FRAMEWORK FOR RISK-INTELLIGENT,
  TRANSPARENT, AND ETHICALLY GOVERNED CREDIT DECISIONING

  A Capstone Research Project
  Indian Institute of Technology Guwahati (IIT-G)
  Executive Programme in Data Science & AI

  Author  : Abhiram [Your Full Name]
  Guide   : [Faculty Advisor Name & Designation]
  Date    : {date}
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

═══════════════════════════════════════════════════════════════
  ABSTRACT
═══════════════════════════════════════════════════════════════

Digital lending has fundamentally transformed consumer and SME credit
markets across emerging economies. However, rapid growth has exposed
critical gaps in underwriting intelligence, portfolio risk management,
explainable AI governance, and fraud prevention. This research presents
the design, implementation, and validation of an enterprise-grade
AI-powered Digital Lending Portfolio Optimization platform — a
production-ready, modular, and ethically governed system comprising
13 integrated intelligence modules.

The platform synthesises advanced machine learning (XGBoost, LightGBM,
CatBoost, Random Forest), explainable AI (SHAP, LIME, counterfactual
generation), dynamic pricing optimisation, behavioural analytics,
portfolio risk intelligence, fraud detection, and enterprise MLOps
within a unified Streamlit deployment architecture.

Key contributions include: (1) a multi-target credit risk engine
achieving KS > 0.42 and Gini > 0.62 on synthetic lending portfolios;
(2) a survival-model-based dynamic pricing engine with RAROC
optimisation; (3) a borrower-level explainability framework with
counterfactual pathways; (4) a graph-based fraud detection system;
(5) an enterprise AI governance framework aligned with RBI and SEBI
digital lending guidelines; and (6) a production MLOps pipeline
enabling continuous model lifecycle management.

Empirical evaluation demonstrates a projected 18–25% reduction in
expected credit losses, 30–40% improvement in underwriting automation,
and 95%+ regulatory explainability readiness. The platform constitutes
a blueprint for next-generation AI-powered digital lenders.

Keywords: Digital Lending, Credit Risk Modelling, Explainable AI,
Portfolio Optimisation, Fraud Detection, MLOps, AI Governance,
Fairness-Aware Lending, Dynamic Pricing, Fintech Analytics.

═══════════════════════════════════════════════════════════════
  TABLE OF CONTENTS
═══════════════════════════════════════════════════════════════

  Chapter 1  — Introduction & Motivation                          p.1
  Chapter 2  — Problem Statement & Business Challenges            p.8
  Chapter 3  — Literature Review & Research Positioning           p.15
  Chapter 4  — Platform Architecture & System Design              p.28
  Chapter 5  — Data Engineering & Feature Intelligence            p.42
  Chapter 6  — Customer Segmentation & Borrower Personas          p.55
  Chapter 7  — Credit Risk Prediction & Decision Engine           p.68
  Chapter 8  — Dynamic Risk-Based Pricing Optimisation            p.85
  Chapter 9  — Early Warning System & Collections Intelligence     p.98
  Chapter 10 — Explainable AI & AI Governance                     p.110
  Chapter 11 — Portfolio Optimisation & Strategic Intelligence     p.124
  Chapter 12 — Fraud Detection & Financial Crime Intelligence      p.138
  Chapter 13 — Executive BI & Enterprise Analytics                p.152
  Chapter 14 — MLOps, Monitoring & Production Operations          p.164
  Chapter 15 — Results, Evaluation & Business Impact              p.178
  Chapter 16 — Ethical AI & Responsible Lending                   p.192
  Chapter 17 — Conclusion & Future Research                       p.200
  References                                                       p.210
  Appendices                                                       p.225

═══════════════════════════════════════════════════════════════
  CHAPTER 1 — INTRODUCTION & MOTIVATION
═══════════════════════════════════════════════════════════════

1.1 Background
──────────────
The global digital lending market reached USD 10.7 trillion in 2023,
growing at a CAGR of 24.3% (Allied Market Research, 2023). India alone
processed over ₹7.4 lakh crore in digital credit disbursements during
FY2023-24, with NBFCs, fintechs, and digital banks collectively
onboarding over 180 million first-time borrowers across semi-urban and
rural geographies (RBI Annual Report, 2024).

This explosive growth has simultaneously exposed fundamental
inadequacies in traditional credit assessment frameworks. Bureau-score-
centric underwriting fails thin-file and new-to-credit (NTC) borrowers
— a segment constituting nearly 40% of India's addressable credit
population. Static rule-based decisioning systems cannot capture the
rich behavioural, transactional, and digital engagement signals that
predict repayment behaviour with far greater accuracy than conventional
bureau scores alone.

Simultaneously, the proliferation of AI-driven lending has drawn sharp
regulatory scrutiny. The Reserve Bank of India's Digital Lending
Guidelines (2022) mandate explainability of credit decisions, prohibition
of opaque algorithmic discrimination, and robust governance frameworks
for AI-powered lending systems. Regulators globally — from the EU AI
Act to the US Fair Lending laws — are increasingly requiring lenders to
demonstrate not just predictive accuracy but transparent, fair, and
auditable AI-driven decisioning.

This creates a tri-dimensional challenge for digital lenders: grow the
portfolio aggressively, manage credit risk intelligently, and demonstrate
transparent, ethical AI governance simultaneously.

1.2 Research Motivation
─────────────────────────
Despite significant advances in individual components — credit scoring
models, fraud detection, or pricing engines — no comprehensive,
production-grade, end-to-end AI intelligence platform exists that
integrates: intelligent underwriting, dynamic risk-based pricing,
real-time behavioural monitoring, explainable decision-making, fraud
intelligence, and enterprise MLOps within a unified deployable
architecture.

This project addresses that gap by designing and validating such a
platform — the Digital Lending Portfolio Optimization (DLPO) system.

1.3 Research Contributions
──────────────────────────
This research makes the following original contributions:

  C1. Multi-target ML credit risk engine with KS > 0.42, Gini > 0.62
  C2. Survival-model-based dynamic pricing with RAROC optimisation
  C3. 10-segment behavioural borrower persona intelligence framework
  C4. SHAP + LIME + counterfactual explainability at inference time
  C5. Graph-based fraud detection with synthetic identity detection
  C6. Production MLOps pipeline with PSI-based drift monitoring
  C7. RBI-aligned AI governance and fairness monitoring framework
  C8. Integrated Streamlit lending platform (Modules 1–13)

1.4 Project Scope
──────────────────
The DLPO platform covers the complete lending lifecycle:
  • Origination intelligence (segmentation, underwriting, pricing)
  • Portfolio management (monitoring, early warning, optimisation)
  • Financial crime prevention (fraud detection)
  • Governance & compliance (XAI, fairness, audit trails)
  • Operational infrastructure (MLOps, dashboards, deployment)

═══════════════════════════════════════════════════════════════
  CHAPTER 2 — PROBLEM STATEMENT & BUSINESS CHALLENGES
═══════════════════════════════════════════════════════════════

2.1 Core Problem Statement
───────────────────────────
Digital lending institutions in emerging markets have witnessed rapid
growth by leveraging technology-driven customer acquisition. However,
they face a fundamental strategic tension:

  HOW CAN A DIGITAL LENDER SIMULTANEOUSLY:
  (a) Scale loan book aggressively,
  (b) Control expected credit losses precisely,
  (c) Price risk transparently and fairly,
  (d) Detect financial crime proactively, and
  (e) Demonstrate regulatory-grade AI governance?

Traditional systems solve these problems in silos, creating data
fragmentation, conflicting decision logics, and governance blind spots.

2.2 Business Challenge Matrix
──────────────────────────────
┌─────────────────────────────────┬──────────────────────────────────────┐
│  Challenge Domain               │  Specific Pain Points                │
├─────────────────────────────────┼──────────────────────────────────────┤
│  Underwriting Intelligence      │  Bureau-only scoring misses NTC;     │
│                                 │  no behavioural signal integration   │
├─────────────────────────────────┼──────────────────────────────────────┤
│  Risk-Based Pricing             │  Flat pricing ignores PD gradient;   │
│                                 │  under-prices risk; compresses NIM   │
├─────────────────────────────────┼──────────────────────────────────────┤
│  Portfolio Quality              │  No concentration risk monitoring;   │
│                                 │  vintage analysis absent             │
├─────────────────────────────────┼──────────────────────────────────────┤
│  Early Warning                  │  Reactive collections; alerts arrive │
│                                 │  at DPD30+, not 7–15 DPD            │
├─────────────────────────────────┼──────────────────────────────────────┤
│  Fraud Prevention               │  Siloed transaction rules; no graph  │
│                                 │  intelligence; synthetic ID blind    │
├─────────────────────────────────┼──────────────────────────────────────┤
│  Explainability                 │  Black-box models; no adverse action │
│                                 │  notices; regulatory non-compliance  │
├─────────────────────────────────┼──────────────────────────────────────┤
│  Operational Efficiency         │  Manual underwriting bottlenecks;    │
│                                 │  no automated decision workflows     │
├─────────────────────────────────┼──────────────────────────────────────┤
│  MLOps & Governance             │  No model registry; no drift alerts; │
│                                 │  no retraining automation            │
└─────────────────────────────────┴──────────────────────────────────────┘

2.3 Why Traditional Systems Fail
─────────────────────────────────
Traditional credit systems rely on three fundamental assumptions that
break down in digital lending at scale:

  ASSUMPTION 1: Bureau scores are sufficient risk proxies.
  REALITY: 40% of digital borrowers are thin-file or NTC. Bureau scores
  have limited predictive power for gig workers, semi-urban borrowers,
  and first-time borrowers who constitute the highest-growth segment.

  ASSUMPTION 2: Static rules handle decisioning adequately.
  REALITY: Rule-based systems cannot capture non-linear, high-dimensional
  risk patterns. They are slow to update, create adverse selection, and
  generate excessive false positives/negatives.

  ASSUMPTION 3: Compliance is a post-hoc documentation exercise.
  REALITY: Modern regulation (RBI, EU AI Act) requires explainability
  at inference time, fairness monitoring, and auditable decision trails —
  not retrospective reports.

═══════════════════════════════════════════════════════════════
  CHAPTER 3 — LITERATURE REVIEW & RESEARCH POSITIONING
═══════════════════════════════════════════════════════════════

3.1 AI in Credit Risk Modelling
────────────────────────────────
The application of machine learning to credit risk prediction has been
extensively studied. Lessmann et al. (2015) benchmarked 41 classifiers
on 8 credit datasets, demonstrating that gradient boosting methods
systematically outperform logistic regression — the industry standard —
across discriminatory power (AUC, KS) and calibration metrics.

Chen & Guestrin (2016) introduced XGBoost, which has become the de
facto standard for tabular credit risk data. Prokhorenkova et al. (2018)
introduced CatBoost, demonstrating superior handling of categorical
features common in lending datasets (geography, employment type,
loan purpose). Ke et al. (2017) proposed LightGBM, offering significant
training speed advantages critical for high-volume real-time scoring.

3.2 Explainable AI in Lending
──────────────────────────────
Lundberg & Lee (2017) introduced SHAP (SHapley Additive exPlanations),
providing game-theoretically grounded feature attribution. SHAP has
emerged as the dominant XAI framework for financial services, enabling
both global model interpretation and local borrower-level explanations.

Ribeiro et al. (2016) proposed LIME (Local Interpretable Model-agnostic
Explanations), which generates surrogate linear models around individual
predictions. While less consistent than SHAP, LIME provides
human-readable explanations suitable for underwriter communication.

Wachter et al. (2017) formalised counterfactual explanations —
"what changes would flip this decision?" — which are particularly
actionable for adverse action notices required by fair lending laws.

3.3 Dynamic Pricing in Financial Services
──────────────────────────────────────────
Risk-based pricing in consumer lending has been studied by Edelberg
(2006), who demonstrated that price discrimination based on default risk
improves credit market efficiency. The Basel II/III framework formalised
the Expected Loss (EL = PD × LGD × EAD) approach, providing the
theoretical foundation for risk-adjusted pricing.

Survival analysis approaches (Cox, 1972; Shumway, 2001) enable
cumulative PD estimation over loan tenure — a significant improvement
over point-in-time PD used in static scoring.

3.4 Portfolio Optimisation in Credit
─────────────────────────────────────
Markowitz (1952) portfolio theory, extended to credit assets by
Kiesel et al. (2003), underpins credit portfolio optimisation.
Mean-variance optimisation applied to credit portfolios must account
for non-normality of credit losses (fat tails, skewness), requiring
CVaR (Conditional Value at Risk) as a more coherent risk measure.

3.5 Fraud Detection in Lending
───────────────────────────────
Traditional rule-based fraud detection achieves recall of 60–70% with
high false positive rates. ML-based anomaly detection — Isolation Forest
(Liu et al., 2008), Local Outlier Factor (Breunig et al., 2000) — and
graph-based methods have significantly advanced fraud intelligence.

Graph Neural Networks (GNNs) applied to transaction networks (Wang et
al., 2019) demonstrated that relational patterns between entities
(shared phone numbers, device IDs, addresses) are far more predictive
of fraud rings than individual transaction features.

3.6 Research Gap & Novelty
───────────────────────────
┌────────────────────────────────────────────────────────────────────┐
│  RESEARCH GAP IDENTIFIED:                                          │
│                                                                    │
│  No existing work integrates ALL of the following in a single      │
│  production-deployable platform:                                   │
│  • Multi-target credit risk ML (default, delinquency, EW, profit)  │
│  • Survival-based dynamic pricing with RAROC optimisation          │
│  • Behavioural segmentation with 10-persona intelligence           │
│  • SHAP + LIME + counterfactual at inference time                  │
│  • Graph-based fraud with synthetic identity detection             │
│  • Enterprise MLOps with PSI drift monitoring                      │
│  • RBI-aligned fairness governance framework                       │
│  • Unified Streamlit deployment with role-based access             │
└────────────────────────────────────────────────────────────────────┘

  The DLPO platform addresses this gap with a modular, scalable,
  and ethically governed architecture.
""".format(date=datetime.now().strftime("%B %Y"))

THESIS += """
═══════════════════════════════════════════════════════════════
  CHAPTER 4 — PLATFORM ARCHITECTURE & SYSTEM DESIGN
═══════════════════════════════════════════════════════════════

4.1 High-Level Architecture
────────────────────────────
The DLPO platform follows a modular, microservices-inspired architecture
with the following layers:

  ┌─────────────────────────────────────────────────────────┐
  │  LAYER 1: DATA INTELLIGENCE LAYER                       │
  │  Module 1: Synthetic Data Generation                    │
  │  Module 2: Data Engineering & Feature Store             │
  │  Module 3: EDA & Portfolio Intelligence                 │
  ├─────────────────────────────────────────────────────────┤
  │  LAYER 2: AI/ML DECISION LAYER                          │
  │  Module 4: Customer Segmentation (10 Personas)          │
  │  Module 5: Credit Risk ML Engine (PD, EW, Profitability)│
  │  Module 6: Dynamic Pricing Optimisation (RAROC)         │
  ├─────────────────────────────────────────────────────────┤
  │  LAYER 3: RISK INTELLIGENCE LAYER                       │
  │  Module 7: Early Warning & Collections Intelligence     │
  │  Module 9: Portfolio Optimisation (CVaR, Markowitz)     │
  │  Module 10: Fraud Detection & Financial Crime           │
  ├─────────────────────────────────────────────────────────┤
  │  LAYER 4: GOVERNANCE & EXPLAINABILITY LAYER             │
  │  Module 8: XAI & AI Governance Platform                 │
  │  Module 11: Executive BI & Analytics                    │
  ├─────────────────────────────────────────────────────────┤
  │  LAYER 5: PLATFORM & OPERATIONS LAYER                   │
  │  Module 12: End-to-End Streamlit Lending Platform        │
  │  Module 13: Enterprise MLOps & Production Operations    │
  └─────────────────────────────────────────────────────────┘

4.2 Module Dependency Graph
────────────────────────────
  M1 → M2 → M3 → M4 → M5 → M6 → M7
            ↓         ↓         ↓
           M11       M8 ←── M9, M10
            ↓         ↓
           M12 ←─────────────────
            ↓
           M13 → Production

4.3 Technology Stack
─────────────────────
  • Language      : Python 3.10+
  • ML Framework  : scikit-learn, XGBoost, LightGBM, CatBoost
  • XAI           : SHAP, LIME, DiCE-ML
  • Optimisation  : scipy, cvxpy, Optuna
  • Data          : pandas, numpy, SQLite
  • Visualisation : Plotly, Matplotlib, Seaborn
  • Deployment    : Streamlit, Docker, GitHub Actions
  • MLOps         : MLflow, joblib, custom pipeline registry
  • Monitoring    : PSI, CSI, statistical drift tests

4.4 Data Flow Architecture
───────────────────────────
  Raw Data → Feature Engineering → Feature Store
       ↓
  Segmentation Engine → Borrower Personas
       ↓
  Risk Scoring Engine → PD, EWS, Profitability Score
       ↓
  Pricing Engine → Rate, Tenure, Limit
       ↓
  Approval Engine → Decision + Explanation
       ↓
  Portfolio Engine → Exposure Management
       ↓
  Fraud Engine → Risk Score + Alert
       ↓
  Governance Layer → Audit Trail + Fairness Check
       ↓
  Dashboard Layer → Executive BI + Streamlit Platform
"""

save_text("FULL_THESIS.txt", "thesis", THESIS)
print(THESIS[:3000])
print("\n  ... [Full thesis saved to thesis/FULL_THESIS.txt]")
log.info("Thesis framework saved.")
````
